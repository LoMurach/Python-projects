```python
import pandas as pd
import pandas_datareader as pdr
```

```python
def prep(symbol_list, start_date, end_date):
    for item in symbol_list:
        stocks = pdr.get_data_tiingo(symbol_list, api_key="83b56978f4990f78e6c167d8886f4510ead7676c", start=start_date, end=end_date)
    df = stocks.reset_index(level=0, drop=False)
    df.index = df.index.date
    df = df.filter(['symbol','adjClose'], axis=1)
    df = df.pivot(columns='symbol', values='adjClose')
    df.columns = sorted(symbol_list, key=lambda L: (L.lower(), L))
    df.to_excel('df.xlsx')
    return df
```

```python
prep(['BTCUSD', 'ETHUSD', 'ETHEUR', 'LTCUSD', 'BTCEUR'], "01-01-2019", "06-30-2019")
```
